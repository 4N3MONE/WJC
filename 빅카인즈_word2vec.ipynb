```python
import pandas as pd
import matplotlib.pyplot as plt
from gensim.models.word2vec import Word2Vec
from konlpy.tag import Okt
```

```python
df_con=df_con.dropna(axis=0)
df_con.isnull().values.any()
```

```python
stopwords=pd.read_csv("C:\\py_temp\\불용어.csv",encoding='utf-8')
stopwords=list(stopwords['단어'])
dd=df_con['content'][:7800]
```

```python
okt=Okt()
sentences_tag=[]
tokens=[]
```

```python
for article in dd:
    temp=okt.morphs(article,stem=True)
    temp = [word for word in temp if not word in stopwords]
    tokens.append(temp)
    # if temp not in stopwords:
    #     tokens.append(temp)
```

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(sentences = tokens,window = 5, min_count = 5, workers = 4, sg = 1)
```

```python
model.wv.vectors.shape
```

```python
result = model.wv.most_similar("서울시",topn=100)
```

```python
resultdf = pd.DataFrame(result, columns=['연관어','유사도'])
resultdf.to_csv('c:\\py_temp\\연관어.csv',encoding="utf-8-sig" , index=False )
resultdf
```
